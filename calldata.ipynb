```python
import pandas as pd
import numpy as np
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
%matplotlib inline

call311data = pd.read_csv("https://data.wprdc.org/datastore/dump/76fda9d0-69be-4dd5-8108-0de7907fc5a4")
call311data.head(10000)
call311data[call311data['REQUEST_TYPE'].str.startswith("Health Hazard")].head(10000)
```

```python
# finding the number of health hazard calls for Squirrel Hill

query_mask = call311data['REQUEST_TYPE'] == "Health Hazard"
query_mask.head(10000)

calls = call311data[query_mask]
calls.head(10000)
calls["NEIGHBORHOOD"].str.startswith("Squirrel").value_counts()
```
